````python
import os, fnmatch

def find(pattern, path): # match a pattern
    result = []
    for root, dirs, files in os.walk(path):
        for name in files:
            if fnmatch.fnmatch(name, pattern):
                print(f'{name}\t{root}')                
                result.append(os.path.join(root, name))
    return result 

path = r'\\che-ds\CS_IgSor\GIT_syno'   
res = find('*.csv', path )  
````

````python
# запись в файл
%%time

with open('/content/purchase_log.txt') as p:
  with open('/content/purchase.txt', 'w', encoding='utf-8') as r:
    
    for i, line in enumerate(p):
      r.write(line)
      
p.close()
r.close()
````

````python
import json
````

````python
%%time

purchase = {}
with open('/content/purchase_log.txt') as p:
    for i, line in enumerate(p):

      c_user = json.loads(line.strip())['user_id']
      c_cat = json.loads(line.strip())["category"]
      
      if c_user in purchase:
        
        purchase[c_user] +=  "-" + c_cat

      else:
        purchase[c_user] = c_cat
      
p.close()

len(purchase)
````

````python
%%time
with open('/content/visit_log.csv') as v:
  counter = 0
  with open('funnel.csv', 'w', encoding='cp1251') as f:
 
    for line in v:
      # если user_id   есть в словаре покупок
      c_user = line.strip().split(',')[0]
      if c_user in purchase:

        #    строка без'\n'   +   значение из словаря 
        f.write(line.strip()+','+ purchase[c_user] +'\n' )
        counter += 1
        
        # для контроля вывода мульти категории
        if counter == 341:  
          print(line.strip()+','+ purchase[c_user] +'\n' )


v.close()    
f.close()

print(f'В файл {f.name} добавлено {counter} строк.')
````
