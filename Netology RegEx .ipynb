**Задание 1**
Напишите функцию, которая принимает на вход строку и проверяет является ли она валидным транспортным номером (1 буква, 3 цифры, 2 буквы, 2-3 цифры). Обратите внимание, что не все буквы кириллического алфавита используются в транспортных номерах.

Если номер валиден, то функция должна возвращать отдельно номер и регион.

Примеры работы программы:

car_id = 'A222BC96'
Результат: Номер A222BС валиден. Регион: 96

car_id = 'АБ22ВВ193'
Результат: Номер не валиден

```python
def plate_checker(text):

    #  ([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})  - A999AA
    #   (1?\d{2})   1(может и не быть)99

    ans = re.findall('^([АВЕКМНОРСТУХ]\d{3}[АВЕКМНОРСТУХ]{2})(1?\d{2})$', text)
    if ans:
        print(f'Номер {ans[0][0]} валиден. Регион: {ans[0][1]}')
        return
    return print('Номер не валиден')
```

```python
plate_checker('A222BC96')
```

```python
plate_checker('АБ22ВВ193')
```

Пример работы программы:

some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'

Результат:
Напишите функцию, которая будет удалять все последовательные повторы слов из заданной строки при помощи регулярных выражений.

```python
def remove_Duplicates(text):

    # \b assert position at a word boundary
    # (\w+) - first Capturing Group
    # \w matches any word character (equivalent to [a-zA-Z0-9_])
    # +  repeat 1+
    # 
    # (?:\W+\1\b)+  Non-capturing group
    #  ?:  - A non-capturing version of regular parentheses
    #  \W matches any non-word character 
    #  +  repeat 1+
    # \1 matches the same text as most recently matched by the 1st capturing group
    # \b assert position at a word boundary 
    
    regex = r'\b(\w+)(?:\W+\1\b)+' 
    return re.sub(regex, r'\1', text)  #, flags=re.IGNORECASE
```

```python
some_string = 'Напишите функцию функцию, которая будет будет будет будет удалять все все все все последовательные повторы слов из из из из заданной строки строки при помощи регулярных выражений.'
```

```python
remove_Duplicates(some_string)
```

**Задание 3**
Напишите функцию, которая будет возвращать акроним по переданной в нее строке со словами.

Примеры работы программы:

some_words = 'Информационные технологии’
Результат: ИТ

some_words = 'Near Field Communication’
Результат: NFC

```python
def make_acronym(text):
    subst = r'\g<1>'
    regex = r'\b\W*([\w])([\w]*\b\W*)' # русский [ЁёА-я]
    result = re.sub(regex, subst, text, 0, re.MULTILINE and re.IGNORECASE)
    if result:
        return result.upper()
```

```python
make_acronym("Информационные технологии  строительства\n")
```

```python
 make_acronym('Near Field Communication')
```

**Задание 4**
Напишите функцию, которая будет принимать на вход список email-адресов и выводить их распределение по доменным зонам.

Пример работы программы:

emails = ['test@gmail.com', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

Результат:

gmail.com: 2
test.in: 1 
ya.ru: 2 
mail.ru: 1

```python
def get_domain(text):
    subst = r'\g<1>'
    regex = r'[A-z0-9!#$%&\'*+-/=?^_`{|}~]*@([A-z0-9][A-z0-9.-]*)'
    result = re.sub(regex, subst, text, 0, re.MULTILINE and re.IGNORECASE)
    if result:
        return result
```

```python
emails = ['test@gmail.com', 'xyz@test.in', 'xyz@test.in', 'xyz@test.in', 'test@ya.ru', 'xyz@mail.ru', 'xyz@ya.ru', 'xyz@gmail.com']

result = {}

for i in emails:
    response = get_domain(i)
    result.setdefault(response,0)
    result[response] += 1

result
```

Напишите функцию, которая будет проверять номер сотового телефона на валидность (только для российского формата), если он валиден, то переводить его в формат:
+7-xxx-xxx-xx-xx
Постарайтесь предусмотреть как можно больше адекватных форматов изначального ввода номера. Примеры работы программы:

phone = '+7 955 555-55-55'
Результат: +7-955-555-55-55

phone = '8(955)555-55-55'
Результат: +7-955-555-55-55

phone = '+7 955 555 55 55'
Результат: +7-955-555-55-55

phone = '7(955) 555-55-55'
Результат: +7-955-555-55-55

phone = '423-555-55-5555'
Результат: Номер не валиден

phone = '123-456-789'
Результат: Номер не валиден

```python
def check_telnumber(text):
    regex = r'\D*[78]?\D*([\d]{3})\D*([\d]{3})\D*([\d]{2})\D*([\d]{2})\D*$'
    subst = r'+7-\g<1>-\g<2>-\g<3>-\g<4>'
    if re.match(regex, text):
        return re.sub(regex, subst, text, 0, re.MULTILINE and re.IGNORECASE)
    else:
        print('Номер не валиден')
```
