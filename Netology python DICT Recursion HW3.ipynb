Дана переменная, в которой хранится словарь, содержащий гео-метки для каждого пользователя (пример структуры данных приведен ниже). Вам необходимо написать программу, которая выведет на экран множество уникальных гео-меток всех пользователей.

Пример работы программы:

ids = {'user1': [213, 213, 213, 15, 213], 
       'user2': [54, 54, 119, 119, 119], 
       'user3': [213, 98, 98, 35]}

```python
ids = {'user1': [213, 213, 213, 15, 213], 'user2': [54, 54, 119, 119, 119], 'user3': [213, 98, 98, 35]}

res = []
for tag in ids.values():
  res += tag

# оставляем уникальные
ans = set(res)  
print(ans)
```

Пример работы программы:

queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
Результат:


Поисковых запросов, содержащих 2 слов(а): 42.86%

Поисковых запросов, содержащих 3 слов(а): 57.14%

```python
queries = [
    'смотреть сериалы онлайн',
    'новости спорта',
    'афиша кино',
    'курс доллара',
    'сериалы этим летом',
    'курс по питону',
    'сериалы про спорт',
]
# list of word quantities equals to whitespaces + 1
res = [q.count(' ') + 1 for q in queries]
# set of available unique values
items = set(res)

for item in items:
  print(f'Поисковых запросов, содержащих {item} слов(а): {res.count(item)/len(res):.2%}')
```

### Задание 4
Дана переменная, в которой хранится статистика рекламных каналов по объемам продаж (пример структуры данных приведен ниже). Напишите программу, которая возвращает название канала с максимальным объемом продаж.

Пример работы программы:

stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}


Результат: Максимальный объем продаж на рекламном канале: vk

```python
stats = {'facebook': 55, 'yandex': 115, 'vk': 120, 'google': 99, 'email': 42, 'ok': 98}

max_sales = max(stats.values() )
#   do not forget about  .items() to retrieve a pair (k, v)
for camp, sales in stats.items():
  if sales == max_sales:
    print(f'Максимальный объем продаж на рекламном канале: {camp}')
```

### Задание 5 (необязательно)
Дан список произвольной длины. Необходимо написать код, который на основе исходного списка составит словарь такого уровня вложенности, какова длина исхондого списка.

Примеры работы программы:

my_list = ['2018-01-01', 'yandex', 'cpc', 100]

Результат: {'2018-01-01': {'yandex': {'cpc': 100}}}

my_list = ['a', 'b', 'c', 'd', 'e', 'f']

Результат: {'a': {'b': {'c': {'d': {'e': 'f'}}}}}

```python
my_list1 = ['2018-01-01', 'yandex', 'cpc', 100]
my_list2 = ['a', 'b', 'c', 'd', 'e', 'f']
my_list3 = ['a','b']  
my_list4 = ['a']  

def build_nested_dict( some_list ):
# если последний элемент списка, то возвращаем строку 
  if len( some_list ) == 1:
#  последний элемент списка -> возвращаем именно СТРОКУ (не словарь)
     return some_list[0]

# возвращаем СЛОВАРЬ с key - первый элемент списка
#              и value - оставшийся срез в рекурсии
  return { some_list[0] : build_nested_dict(some_list[ 1 : ] ) }

print(build_nested_dict(my_list1) )
print(build_nested_dict(my_list2) )
print(build_nested_dict(my_list3) )
print(build_nested_dict(my_list4) )
#  вложенная структура словаря формируется при длине списка больше 2-х элементов
```

```python
from functools import reduce
print(my_list1[::-1]) # перевёрнутый список
[100, 'cpc', 'yandex', '2018-01-01']
# берем key, value и формируем словарь {value: key}, который станет значением в новом словаре и  т.д.
# 100,'cpc' -> {'cpc',100}, ..{'cpc',100}, 'yandex', ->  {'yandex': {'cpc': 100}} .. {'yandex': {'cpc': 100}} , '2018-01-01' -> {'2018-01-01':{'yandex': {'cpc': 100}} }
print(reduce(lambda key, value: {value: key}, my_list1[::-1]) )
# print(reduce(lambda key, value: {value: key}, my_list1) )
```

```python
import re

cook_book = {
  'салат': [
     {'ingridient_name': 'сыр', 'quantity': 50, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 2, 'measure': 'шт'},
     {'ingridient_name': 'огурцы', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'маслины', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'оливковое масло', 'quantity': 20, 'measure': 'мл'},
     {'ingridient_name': 'салат', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'перец', 'quantity': 20, 'measure': 'гр'}
    ],
  'пицца': [
     {'ingridient_name': 'сыр', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'колбаса', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'бекон', 'quantity': 30, 'measure': 'гр'},
     {'ingridient_name': 'оливки', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'томаты', 'quantity': 20, 'measure': 'гр'},
     {'ingridient_name': 'тесто', 'quantity': 100, 'measure': 'гр'},   
    ],
  'лимонад': [
     {'ingridient_name': 'лимон', 'quantity': 1, 'measure': 'шт'},
     {'ingridient_name': 'вода', 'quantity': 200, 'measure': 'мл'},
     {'ingridient_name': 'сахар', 'quantity': 10, 'measure': 'гр'},
     {'ingridient_name': 'лайм', 'quantity': 20, 'measure': 'гр'},    
    ]
}

pc = int(input('Введите количество порций: '))
print()
# инициализируем словарь ингридиентов
ingr_dict = {}

# перебираем блюда
for meal, info in cook_book.items():
  # перебираем состав блюд
  for item in info:
    ingr, q, m = item['ingridient_name'],item['quantity'],item['measure']

    # просмотр словаря ингридиентов
    if ingr + ' (' + m + ')' in ingr_dict: 
      # если в словаре есть сочетание ингридиента и ед.измерения,
      # то добавляем количество
      ingr_dict[ingr + ' (' + m + ')'] += q * pc
    else:
      # добавляем новый ключ (по типу 'бекон (гр)') и значение (количество)
      ingr_dict[ingr + ' (' + m + ')'] = q * pc 

# выводим список ингридиентов   
for meal, info in sorted(ingr_dict.items()):
  print (f'{meal.capitalize()}: {info}')

print('\nПодгонка под формат вывода в условии задания:\n')
for meal, info in ingr_dict.items():
  name = re.search('.+[ ]', meal).group(0)
  measure = re.search('[\(](.+)[\)]', meal).group(1)
  print (f'{name.capitalize().rstrip()}: {info} {measure}')  
```
