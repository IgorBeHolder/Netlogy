### Задание 1


Посчитайте значение первой и второй производной функции:

$f(x) = x^5 + 4sin(2x) + cos(3x+3)$ 

В точке x = 1.


```python
### YOUR CODE HERE
import sympy as sp
from sympy.plotting import plot
import numpy as np
import matplotlib.pyplot as plt

x, y, z = sp.symbols('x y z')
```

```python
# f = sp.Function('f')
# f(x)
```

```python
# the function to differetiate
func= x**5 + 4 * sp.sin(2 * x)  + sp.cos(3 * x + 3)
func
```

```python
# first derivative with respect to x in algebraic form
sp.diff(func,x,1)
```

```python
# second derivative with respect to x in algebraic form
sp.diff(func,x, 2)
```

```python
first_derivative = sp.diff(func,x,1)
second_derivative = sp.diff(func,x,2)
res1 = first_derivative.subs(x, 1)
real_res1 = res1.evalf()
res2 = second_derivative.subs(x, 1)
real_res2 = res2.evalf()
print(f"f(x)={func}")
print(f"f'(x) = {res1},    f'(1)= {real_res1}")
print(f'f"(x) = {res2}, f\'(1)= {real_res2}')
```
