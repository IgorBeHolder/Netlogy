```python
class RateHacked (Rate):
    def __init__(self):
        super().__init__(format = 'value')
        self.diff = False
    def make_format(self, currency):
        """
        Возвращает информацию о валюте currency в двух вариантах:
        - полная информация о валюте при self.format = 'full':
        Rate.make_format('EUR'):
        {
            'CharCode': 'EUR',
            'ID': 'R01239',
            'Name': 'Евро',
            'Nominal': 1,
            'NumCode': '978',
            'Previous': 79.6765,
            'Value': 79.4966
        }
        - сжатая информация - выводится только текущий курс валюты
        Rate.make_format('EUR'):
        79.4966

        - при значениие self.diff = True выводится изменения курса

        
        """
        response = self.exchange_rates()
        # если валюта есть в ответе
        if currency in response:
            # и формат отчета полный
            if self.format == 'full':
                # dthyb gjkysq jnxtn
                return response[currency]
            # если отчет только по валюте
            if self.format == 'value':
                # верни отчет по валюте (если не diff)              иначе #  верни разницу валют
                return response[currency]['Value']  if not self.diff else response[currency]['Value']-response[currency]['Previous']
        
        return 'Error'        
```

```python
mr = RateHacked() #  инициализируем класс
data = mr.exchange_rates() # запрос данных
```

```python
# iterate through VALUES, extract ['Value'] and yield max
max_value = max(data.values(), key=lambda x: x['Value'])
# extract Name field value
print(f'Курс {max_value["Name"]}: {max_value["Value"]}')
```

**Комментарий** по классу Designer такой:
Напишите класс Designer, который учитывает количество международных премий для дизайнеров (из презентации: “Повышение на 1 грейд за каждые 7 баллов. Получение международной премии – это +2 балла”). Считайте, что при выходе на работу сотрудник уже имеет две премии и их количество не меняется со стажем (конечно если хотите это можно вручную менять).

Класс Designer пишется по аналогии с классом Developer из материалов занятия.

```python
class Employee:
    def __init__(self, name, department):
        self.name = name
        self.department = department
        
class Designer (Employee):
    def __init__(self, name, department, int_award=2):
        super().__init__(name, department)    
        self.int_award = int_award
        self.grade = 0
    def get_int_award(self, point):
        self.int_award += point
        self.check_grade_status()

    def check_grade_status(self):
        if self.int_award / 7 >= self.grade:
            self.grade = self.int_award // 7
```

```python
c = Designer('Елизавета Игоревна', 'e-media', 3)
```
