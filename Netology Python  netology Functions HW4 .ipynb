```python
documents = [
 {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
 {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
 {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]
directories = {
 '1': ['2207 876234', '11-2'],
 '2': ['10006'],
 '3': []
}
```

Инициализация

```python
import time

def main():
  command_list()
  input_task()

def command_list():
  print('l - список документов')
  print('ls - список полок')  
  print('p - поиск документа по номеру')
  print('s - поиск полки по номеру')
  print('ad - добавление нового документа')
  print('d - удаление документа')  
  print('m - перемещение документа на другую полку')    
  print('ads - добавление новой полки')
  print('ds - удаление не пустой полки')  
  print('h - список команд')
  print('q - завершение работы')  
  

def input_task():
  tic = time.perf_counter()
  task = ''

  while task !='q':
    task = input('Введите команду: ')
    if task == 'p':
      docnum_input()
    elif task == 's':
      shelf_search()
    elif task == 'm':
      doc_move()    
    elif task == 'l':
      print_doclist()     
    elif task == 'ads':
      add_shelf()          
    elif task == 'ds':
      delete_shelf()         
    elif task == 'h':
      command_list() 
    elif task == 'ls':         
      print_shelflist()  
    elif task == 'ad':
      add_document() 
    elif task == 'd':         
      delete_document()         
    toc = time.perf_counter()                
  return print(f'Сеанс работы завершен.\nВремя сеанса {toc - tic:0.1f} сек')
```

Операции с документами

```python
def docnum_input(add = False):  
  num = input('Введите номер документа:')
  result = doc_search(num, add)
  return (result)  
  
def doc_search(num, add):
  for doc in documents:
    if doc['number'] == num:
        print(f'Владелец документа: {doc["name"]}')
        return (num, doc['name'])
  else:
    if not add:
      print(f'Документ не найден в базе.\n') 
      print_doclist()
    else:
      # добавление нового документа -> возвращаем введенный номер
      return(num,'')

  return ('','')    

def print_doclist():
    print(f'Текущий список документов:')
    for doc in documents:
      for shelf, doc_list in directories.items():
        if doc["number"] in doc_list:
          cur_shelf = shelf
          print(f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {cur_shelf}')

def doc_move():
  num, name = docnum_input() 
  if num != '':
    # если документ найден, выбираем полку
    new_shelf = shelf_choice()  
    if new_shelf ==0:
      # если полка не найдена
      return
  else: 
    # если документ не найден
    return
  
  # перемещаем документ
  for fold, item in directories.items():
    if num in item:
      # удаляем из старой полки
      item.remove(num)
    if fold == new_shelf:
      # добавляем на новую полку
      item.append(num)

  print(f'Документ перемещен на: {new_shelf} полку')
            

def add_document():
  num, name = docnum_input(add = True) 
  if num != '' and name != '':
    # если документ найден
    print('Документ уже существует')
  else: 
    # добавление документа
    doc_type = input('Введите тип документа:')
    doc_name = input('Введите владельца документа:')
    doc_shelf = 0
    while doc_shelf == 0:
      doc_shelf = shelf_choice()

    documents.append({'type': doc_type, 'number': num, 'name': doc_name})
    directories[doc_shelf].append(num)
    print('Документ добавлен')
    print_doclist()
  


def delete_document():
  num,name = docnum_input(add =False)
  for doc in documents:
    if doc['number'] == num:
       documents.remove(doc)
       print(f'Документ удален')
       print_doclist()
   
```

Операции с полками

```python
def shelf_search():
  num, name = docnum_input() 
  for shelf, item in directories.items():
    if num in item:
      print(f'Документ хранится на полке: {shelf}')
      return
  print('Документ не найден в базе')
  

def shelf_choice():  
  shelf = input('Введите номер полки:')
  if shelf in directories:
    return shelf
  else:
    print(f'Такой полки не существует. \n')
    print_shelflist()
  return 0
  
def add_shelf():  
  new_shelf = input('Введите номер полки:')
  if new_shelf in directories:
    print('Такая полка уже существует')
  else:
    directories[new_shelf] = []
    print('Полка добавлена')
    print_shelflist()


def print_shelflist():
  print(f'Текущий перечень полок: { ", ".join([x for x in directories.keys()]) }')  

def delete_shelf():    
  shelf_todelete = shelf_choice() 
  if len(directories[shelf_todelete]) == 0:
    directories.pop(shelf_todelete)
  else:
    print('На полке есть документы, удалите их перед удалением полки.')
  print_shelflist()
  
```

Запуск

```python
main()  
```
