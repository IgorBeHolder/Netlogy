**Задание 1**
Печатные газеты использовали свой формат дат для каждого выпуска. Для каждой газеты из списка напишите формат указанной даты для перевода в объект datetime:
The Moscow Times - Wednesday, October 2, 2002
The Guardian - Friday, 11.10.13
Daily News - Thursday, 18 August 1977

```python
from datetime import datetime,timedelta
# справочник форматов дат
date_formats = {
    'Monday, January 01, 2022' : '%A, %B %d, %Y',
    'Monday, 01.12.22' : '%A, %d.%m.%y',
    'Monday, 01 January 2022' : '%A, %d %B %Y',
    'YYYY.MM.DD' : '%Y.%m.%d',
    'YYYY-MM-DD' : '%Y-%m-%d',
}

def check_date(date, format):
    '''
    Return True if date and format are a valid combination
    '''
    try:
        datetime.strptime(date, format)
        return True
    except:
        return False

def seek_dateformat(date):
    '''
    Return available output python date format
    '''
    # есть ли формат, соответствующий date
    for k, df in date_formats.items():
        if check_date(date, df):
            return df
    else:
        # print(f'Unknown date_format for {date}.') 
        return '%Y.%m.%d'

def date_format(date, cur_format = None):
    ''' 
    Convert date to cur_format ('%Y.%m.%d if None)
    '''
    if cur_format is None:
        if check_date(date, seek_dateformat(date)):
            return datetime.strftime(datetime.strptime(date, seek_dateformat(date)), '%Y.%m.%d')
    else:
        if check_date(date, seek_dateformat(date)):
                if cur_format in date_formats:
                    return datetime.strftime(datetime.strptime(date, seek_dateformat(date)), date_formats[cur_format])
                else:
                    print(f'Format {cur_format} is not in date_format list or not valid.')   
        else:
            # print(f'Unknown date_format for {date}.') 
            return 
# все строки прeобразуются в питоновский формат даты, затем происходит вывод в output формат
```

```python
inp ='The Moscow Times - Wednesday, October 2, 2002 - The Guardian - Friday, 11.10.13 - Daily News - Thursday, 18 August 1977'
# формируем список из входной строки
list_ = inp.strip().split('-')
print(f'List to convert: {list_}\n')
result = {}
paper = list_[0].strip() # название первой газеты
result.setdefault(paper) 

for i, l in enumerate(list_):


    item = l.strip()
  
    if i % 2 != 0:  #  если нечетное число
        # можно указать любой выходной формат, по умолчанию 'YYYY.MM.DD'
        result[paper]= date_format(item)  
        # result[paper]= date_format(item, 'Monday, 01.12.22')   
    else:
      result.setdefault(item)
      paper = item

result
```

**Задание 3**

Напишите функцию date_range, которая возвращает список дат за период от start_date до end_date. Даты должны вводиться в формате YYYY-MM-DD. В случае неверного формата или при start_date > end_date должен возвращаться пустой список.

```python
def check_date_input(date, date_format):
    if check_date(date, date_format):
        return True
    else:
        return False

def  date_range(start_date, end_date):
    if   start_date > end_date:
        return list()
    else:
        
        return [datetime.strftime(start_date + timedelta(days=d), '%Y-%m-%d')
                                 for d in range((end_date - start_date).days+1)]

def date_input(str, date_format):
    while True:
        d = input(f'Enter {str} ({date_format}):')
        try:
            if check_date_input(d, date_formats[date_format]):
                return d
        except:
            print(f'Unknown date_format for {d, date_format}.') 
            return 
             
```

```python
date_format = 'YYYY-MM-DD' # задаем формат вывода по умолчанию  'YYYY-MM-DD'
start_date = datetime.strptime(date_input('Start date', date_format), date_formats[date_format]) 
end_date = datetime.strptime(date_input('End date', date_format), date_formats[date_format]) 
date_range(start_date, end_date)#, date_format)
```
