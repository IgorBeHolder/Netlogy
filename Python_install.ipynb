<h4><code>!pip install</code> vs <code>%pip install</code></h4>
<p>Any command prefixed with <code>!</code> is treated as a shell command in Jupyter cells. Thus <code>!pip install &lt;module&gt;</code> is treated as a simple shell command that translates to <code>pip install &lt;module&gt;</code>. However the recommendation is to use <code>python -m pip install &lt;module&gt;</code><h4>  

To get this desired behavior one must use  
## <code>%pip install &lt;module&gt;</code></p>
<h4><code>%conda install</code><h4>
<p>However when installing packages in Jupyter into a conda environment,  
use of <code>conda install</code> is preferred over <code>pip install</code>. Hence its highly recommended that one use <code>%conda install</code> in jupyter notebooks when dealing with Conda enviornments.</p>

```python
'''each Python executable has its own site-packages: 
what this means is that when you install a package, it is associated with particular python executable 
and by default can only be used with that Python installation!
We can see this by printing the sys.path variables for each of the available python executables in my path, 
using Jupyter's delightful ability to mix Python and bash commands in a single code block:'''
paths = !type -a python3
for path in set(paths):
    path = path.split()[-1] # get the last word
    print(path)
    !{path} -c "import sys; print(sys.path)"
    print()
```

### import python packages

```python
soft_list = {'pandas': 'pd', 'numpy': 'np'}             # list of packages to import
for  package in soft_list.keys():                       # loop over the list
    exec(f'import {package} as {soft_list[package]}')   # import the package
    print(f'{package} imported as {soft_list[package]}  version: {eval(soft_list[package]).__version__}')
```

```python
%%time
import sys
import os
# install a pip package in the current Jupyter kernel
soft_list = ['pandas',
             'numpy',
             'matplotlib',
             'plotly.express',
            #  'nbformat',
             'seaborn',
            #  'pymorphy2',
             'scikit-learn',
             'scipy',
            #  'nltk',
            #  'gensim',
            #  'torch',
            #  'tensorflow',
             'bs4',
             'selenium',
             'requests',
             'folium',
            #  'geopy',
            #  'openpyxl',
            #  'cookiecutter',
            #  'wordcloud',
            #  'opencv-python',
            #  'skimage'
             ]

for package in soft_list:
    !{sys.executable} -m pip install {package} -q
os.say('All packages installed')
```
